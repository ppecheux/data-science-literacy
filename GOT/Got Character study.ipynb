```python
with open('GOT-characters-raw.pickle','rb') as fp:
    characters =pickle.load(fp)
len(characters.keys())
```

How many names do we have?

```python
entities=0
for v in characters.values():
    entities+=len(v)
print(entities)
```

How many unic name?

```python
entities=set()
for v in characters.values():
    for e in v:
        name = e[0]
        entities.add(name)
print(len(entities))
```

```python
GOT=nx.Graph()
GOT.add_nodes_from(list(entities))
print(nx.info(GOT))
```

```python
chapter = characters['1–1-Bran']
chapter.sort(key=lambda a:a[1],reverse=False)
print(chapter)
```

```python
#donne tous les liens possibles entre les personnages
potential_edges=[]
for i in range(0,len(characters['1–1-Bran'])):
    name1,start1,end1 = characters['1–1-Bran'][i]
    for j in range(i+1,len(chapter)):
        name2,start2,end2 = chapter[j]
        if name2 != name1:
            potential_edges.append((name1,name2,{'dist':start2-end1}))
    
print(len(potential_edges))
subG=nx.Graph()
subG.add_edges_from(potential_edges)
```

```python
print(nx.info(subG))
```

```python
#garde un seul couple personnage avec la liste des distances entre les deux
edges = {}
for name1,name2,value in potential_edges:
    if (name1,name2) not in edges.keys():
        if (name2,name1) not in edges.keys():
            edges[(name1,name2)] = {'values':[value['dist']]}
        else:
            edges[(name2,name1)]['values'].append(value['dist'])
    else:
        edges[(name1,name2)]['values'].append(value['dist'])

for k,v in edges.items():
    edges[k]['weight']=np.percentile(v['values'],25)
edges
```

```python
for k,v in edges.items():
    edges[k]['weight']=len([a for a in v['values'] if a <1000])
    print(k,edges[k]['weight'])
```

```python
ready_for_nx =[]
for k,v in edges.items():
    name1=k[0]
    name2=k[1]
    weight=v['weight']
    ready_for_nx.append((name1,name2,weight))
return ready_for_nx
```
