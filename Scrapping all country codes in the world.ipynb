```python
from bs4 import BeautifulSoup
```

read html and parse cells line by line

```python
with open ('iso-country-code.html','r')as fp:
    html=fp.read()
    soup=BeautifulSoup(html,'html5lib')
trs= soup.find_all('tr')
len(trs)
```

```python
countries=[]

for tr in trs:
    values=[td.string for td in tr.find_all('td')]
    if values[0] is not None and len(values[0])>0:
        #print(values[0],values[3])
        countries.append((str(values[0]),str(values[3])))
                         
countries=countries[1:-1]
countries
```

```python
import requests
```

```python
def get_indicator(country_code,indicator):
    head = 'http://api.worldbank.org/v2/country/'
    middle = '/indicator/'
    tail = '?format=json'
    url = head+country_code+middle+indicator+tail
    r = requests.get(url)
    try:
        data = r.json()
        try:
            for i in data[1]:
                if i['value'] is None:
                    continue
                else:
                    return i['value']
        except TypeError:
            return None
    except KeyError:
        return None
    except IndexError:
        return None
    except TypeError:
        return None
        
get_indicator('CHN','GC.TAX.IMPT.ZS')
```
