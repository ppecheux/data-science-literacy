## Parsing country codes with BS4

```python
from bs4 import BeautifulSoup #parsing
```

'tr' defines a row in HTML
There is 251 rows in this html file

```python
with open ('iso-country-code.html','r')as fp:
    html=fp.read()
    soup=BeautifulSoup(html,'html5lib')
trs= soup.find_all('tr')
len(trs)
```

The 'td' element represents a data cell in a table. Parsing cell by cell, we get the dictionary country name : code

```python
countries = {}
for tr in trs:
    values = [td.string for td in tr.find_all('td')]
    if values[0] is not None and len(values[0]) > 0:
        countries[str(values[0])] = str(values[3])
print('All the countries done')
del countries['\xa0']
countries
```

```python
with open('all_countries.pickle','wb+')as fp:
    pickle.dump(countries,fp)
```

## fill the indicators
Now that we have all the countries in the B collumn we can fill with the indicators

```python
def get_indicator(country_code,indicator):
    """this founction search for the last infdicator
        for the desired country
        using the world bank api"""
    
    head = 'http://api.worldbank.org/v2/country/'
    middle = '/indicator/'
    tail = '?format=json'
    url = head+country_code+middle+indicator+tail
    r = requests.get(url)
    try:
        data = r.json()
        try:
            for i in data[1]:
                if i['value'] is None:
                    continue
                else:
                    return i['value']
        except TypeError:
            return None
    except KeyError:
        return None
    except IndexError:
        return None
        
get_indicator('CHN','GC.TAX.IMPT.ZS')
```

```python
indicator_list={
'IMF WB\nCustoms and other import duties (% of tax revenue) 2016':'GC.TAX.IMPT.ZS',
'WB IMF Non-performing loans to total gross loans ratio (%)':'FB.AST.NPER.ZS',
'UIC Rail freight (mn ton x km travelled)':'IS.RRS.GOOD.MT.K6',
'WB Household final consumption expenditure, PPP bn USD':'NE.CON.PRVT.PP.CD',
'WB UNESCO R&D Expenditure % of GDP':'GB.XPD.RSDV.GD.ZS',
'WB UNESCO Researchers p. mn inhabitants':'SP.POP.SCIE.RD.P6',
'IFDI Net Inflows (% of GDP)':'BX.KLT.DINV.WD.GD.ZS',
'WHO Current Health Expenditure p.c., PPP USD':'SH.XPD.CHEX.PP.CD',
'WB Research Expenditure % of GDP':'GB.XPD.RSDV.GD.ZS',
'WIPO Trademark applications':'IP.TMK.RSCT',
'WIPO Patent Applications':'IP.PAT.RESD'
}
```

```python
i=1 #because the first country is on the 2 row
for name,code in countries.items():
    print(name+'...'+code+'...')
    for il in indicator_list :
        i+=1
        #try:
        #   if(not pandas.isna(df.at[i,il])):
        #       print(df.at[i,il])
        #       break
        #except KeyError:
        #    print("error line " + str(i))
        value = get_indicator(code,indicator_list[il])
        time.sleep(0.5+random.random())
        if value is not None:
            df.at[i,il]=value
        
    print('done...')
    
df
```
